```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

"""
    file_name = "fz.csv"
    data = pd.read_csv(file_name, sep=";",names=["x1","x2","y"])
    data['x1'] = data['x1'].str.replace('.','',regex=True)
    data['x2'] = data['x2'].str.replace('.','',regex=True)
    data['y'] = data['y'].str.replace('.','',regex=True)
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(data)
                
    # Separamos las columnas de las entradas y salidas
    x = data_scaled[:, :-1]
    y = data_scaled[:, -1]
"""

x = ([[0.12396694, 0.75316456],
       [1.        , 0.82911392],
       [0.35743802, 0.02742616],
       [0.32024793, 0.91983122],
       [0.83057851, 0.59704641],
       [0.20867769, 0.93037975],
       [0.78305785, 0.69620253],
       [0.04752066, 0.02320675],
       [0.41735537, 0.78902954],
       [0.30578512, 0.45147679],
       [0.52479339, 0.10759494],
       [0.91115702, 0.02953586],
       [0.23760331, 0.48101266],
       [0.80785124, 0.06962025],
       [0.50826446, 0.78270042],
       [0.89876033, 0.30168776],
       [0.9731405 , 0.27848101],
       [0.54958678, 0.92616034],
       [0.94008264, 0.51054852],
       [0.01446281, 0.32278481],
       [0.94008264, 0.94725738],
       [0.99173554, 0.56751055],
       [0.95867769, 0.27004219],
       [0.70867769, 0.51898734],
       [0.        , 0.49367089],
       [0.98966942, 0.17932489],
       [0.71487603, 0.76371308],
       [0.94421488, 0.07172996],
       [0.3161157 , 0.94725738],
       [0.34710744, 0.28902954],
       [0.92355372, 0.98734177],
       [0.53305785, 0.28059072],
       [0.95454545, 0.48101266],
       [0.09710744, 1.        ],
       [0.54752066, 0.2742616 ],
       [0.06404959, 0.72995781],
       [0.9338843 , 0.51054852],
       [0.19628099, 0.50632911],
       [0.53305785, 0.72151899],
       [0.95247934, 0.32911392],
       [0.84504132, 0.52109705],
       [0.24173554, 0.22151899],
       [0.54338843, 0.16455696],
       [0.08471074, 0.69831224],
       [0.41942149, 0.43037975],
       [0.0785124 , 0.28902954],
       [0.20867769, 0.70675105],
       [0.7231405 , 0.85864979],
       [0.79958678, 0.2742616 ],
       [0.01239669, 0.43881857],
       [0.00206612, 0.55696203],
       [0.79752066, 0.85232068],
       [0.78305785, 0.01265823],
       [0.71694215, 0.1814346 ],
       [0.16735537, 0.81434599],
       [0.80578512, 0.        ],
       [0.00619835, 0.30168776],
       [0.95247934, 0.74683544],
       [0.71280992, 0.52953586],
       [0.00619835, 0.55063291]])

y = ([0.44495413, 0.96972477, 0.09541284, 0.68348624, 0.81009174,
       0.59724771, 0.83761468, 0.        , 0.69908257, 0.45963303,
       0.30917431, 0.36422018, 0.42293578, 0.38623853, 0.74862385,
       0.65504587, 0.66238532, 0.82385321, 0.80458716, 0.06513761,
       0.99266055, 0.85321101, 0.65045872, 0.72201835, 0.14862385,
       0.57889908, 0.83853211, 0.43944954, 0.68990826, 0.37889908,
       1.        , 0.48348624, 0.79357798, 0.50275229, 0.48440367,
       0.36146789, 0.80275229, 0.40183486, 0.73669725, 0.69449541,
       0.77706422, 0.2412844 , 0.38256881, 0.37522936, 0.52385321,
       0.1293578 , 0.50917431, 0.87889908, 0.59724771, 0.13853211,
       0.18715596, 0.90642202, 0.28807339, 0.47981651, 0.5146789 ,
       0.27431193, 0.03486239, 0.92110092, 0.7293578 , 0.18899083])

def membership_values(x, K):
    output = []
    n_features  = len(K)
    for sample in x:
        membership_i = [] # Grados de pertenencia de una muestra
        for f in range(n_features):
            tmp = [] # Lista temporal --> Almacen de grados de pertenencia para una feature
            for j in range(1, K[f]+1):
                # mu
                a = (j - 1) / (K[f] - 1)
                b = 1 / (K[f] - 1)
                value = 1 - abs(sample[f] - a) / b
            if value < 0:
                tmp.append(0)
            else: 
                tmp.append(value)
        # Añadimos los grados de pertenencia de la característica a membership_i
        membership_i.append(tmp)
    # Lo mismo pero en el output
    output.append(membership_i)
    return output

# Pedro Marti
"""
    - k: num triángulos a generar
    - min_val: valor mínimo del rango en el que trabajamos
    - max_val: valor máximo
"""
def k_partitions(k, min_val, max_val) -> np.ndarray:
    segment_size  = (max_val - min_val) / (k - 1) # Tamaño del segmento
    triangle_points = [] # Coordenadas de los triangulos
    current_point = min_val # Inicializamos contador

    # Empezamos a generar los triangulos
    for i in range(k):
        if i == 0: # Si es el primer triangulo, solo es la mitad
            triangle_points.append([min_val, min_val, min_val + segment_size])
            current_point = min_val + segment_size
        elif i == k -1: #Si es el último triángulo, solo es la mitad
            triangle_points.append([max_val - segment_size, max_val, max_val])
        else: # Cualquier otro caso, es un triángulo completo
            triangle_points.append([current_point - segment_size, current_point, current_point + segment_size])
            current_point = current_point + segment_size

    return np.array(triangle_points)

# Calcular el valor de b para una regla
def calculate_b(x_values, y_values, z, mu):
    numerator = 0
    denominator = 0
    for x, y in zip(x_values, y_values):
        w = mu(x) ** z
        numerator += w * y
        denominator += w
    return numerator / denominator if denominator != 0 else None

# Generate all posible combinations for all the feature K_i
"""
Here we just generate all posible combinations for all the feature k_i for 
example k = [2,2] will return [[1,1],[1,2],[2,1],[2,2]]
"""
def combinations_for_k(k:np.ndarray):
    # List to store combinations
    output = [] 
    num_features = len(k)
    # Calculate total number of combinations
    num_combinations = np.prod(k)

    # Iterate from 0 to num_combinations - 1
    for i in range(num_combinations):
        # Store current combination
        combination = []
        temp_i = i

        # For each K_i
        for j in range(num_features):
            comb_value = temp_i % k[j]
            combination.append(comb_value)
            # Update i for next K_i rounding it to nearest whole number
            temp_i = temp_i // k[j]
    
        output.append(combination)

    return output

#Compute every degree of compatibility with every combination. There are k_1*k2*...k_n different combinations
def degree_of_compatibility(membership_values, combinations):
    final_list = []
    
    # Looping through each membership value
    for mem_val in membership_values:
        temp_list_outer = []
        
        # Looping through each combination
        for comb in combinations:
            temp_list_inner = []
            
            # Looping through each feature
            for idx in range(len(comb)):
                
                # Check if membership_values has only one feature
                if len(mem_val) == 1:
                    temp_list_inner.append(mem_val[0][comb[idx]])
                
                # Else, assume it has multiple features
                else:
                    temp_list_inner.append(mem_val[idx][comb[idx]])
            
            # Convert to numpy array for mathematical operation
            np_temp_list_inner = np.array(temp_list_inner)
            
            # Calculate product and append to outer list
            temp_list_outer.append(np.product(np_temp_list_inner))
        
        # Append each computed list to the final list
        final_list.append(temp_list_outer)
    return final_list



# Peso
def Wj1_jn(dc,k:np.ndarray, alpha=1):
    return np.power(dc,alpha)

"""
b = Variable de salida en función de reglas difusas
"""
def heuristic_method_of_b(weight_matrix, target_vals):
    b_values_list = [] #Lista de resultados
    transposed_matrix = np.transpose(weight_matrix) # Facilita la iteración por columnas

    for column in transposed_matrix:
        # Numerator operation: EW_j1..jn(x_p) * Y_p
        num_sum = np.dot(column, target_vals)

        # Denominator operation: EW_j1..jn(x_p)
        denom_sum = np.sum(column)

        # Calculamos b
        if denom_sum != 0:
            b_values_list.append(num_sum/denom_sum)
        else:
            b_values_list.append(0)
    return np.array(b_values_list)

# Predict
def predict(input_data, coeff_b, coeff_B, coeff_k):

    # Cálculo de valores de pertenencia
    memb_values_b = membership_values(coeff_b, coeff_B)
    memb_values_input = membership_values(input_data, coeff_k)

    # Generar las posibles combinaciones y calculamos su grado de compatibilidad
    possible_combinations = combinations_for_k(coeff_k)
    dc_values = degree_of_compatibility(memb_values_input, possible_combinations)

    predict_results = []
    primary_memb_table = np.zeros((coeff_k[0], coeff_k[1])) # coeff_k[0] = 3, coeff_k[1] = 3
    secondary_memb_table = np.zeros((coeff_k[0], coeff_k[1]))

    # Iteramos sobre todas las posibles combinaciones de características
    for i, combination in enumerate(possible_combinations):
        if len(memb_values_b) == 0:
            print("memb_values_b esta completamente vacíoS")
        elif len(memb_values_b[i]) == 0:
            print(f"Cuidado: memb_values_bb[{i}] esta vacia")
        elif len(memb_values_b[i]) > 0:
            primary_idx = np.argmax(memb_values_b[i])
        else:
            print(f"No se ha encontrado argmax para una secuencia vacía en el índice {i}")

        print(f"memb_values_b[i]: {memb_values_b[i]}") # DEBUG
        primary_idx = np.argmax(memb_values_b[i]) # Buscamos indice de al etiqueta con mayor pertenencia
        primary_memb_table[combination[0]][combination[1]] = primary_idx 

        secondary_values = memb_values_b[i].copy()
        secondary_values[primary_idx] = -1 # Anulamos el valor más alto para encontrar el segundo más alto
        secondary_idx = np.argmax(secondary_values) # Encontramos el segundo valor más alto
        secondary_memb_table[combination[0]][combination[1]] = secondary_idx
    
    """ 
    Realizamos la predicción utilizando las tablas y grados de
    .compatibilidad, iterando sobre cada muestra.
    """
    for sample in dc_values: # Sample = Vector con los grados de compatibilidad de esa muestra con cala regla posible
        pred = sum(sample * primary_memb_table.flatten()) / sum(sample)
        predict_results.append(pred)
        
    return predict_results, primary_memb_table, secondary_memb_table

def simple_prediction(input_x, input_y, k_values, b_values, alpha = 1.0):
    all_memb_values = membership_values(input_x, k_values)
    comb_values = combinations_for_k(k_values)
    dc_matrix = degree_of_compatibility(all_memb_values, comb_values)
    w_values = np.power(dc_matrix, alpha)
    b_calc = heuristic_method_of_b(w_values, input_y)

    # Realizamos las predicciones
    pred_output, primary_table, secondary_table = predict(input_x, b_calc, b_values, k_values)

    pi_val = np.sum(np.power(np.substract(pred_output, input_y), 2))/2
    print(f"PI = {pi_val}")

    b_table = np.zeros_like(primary_table)

    for idx, comb in enumerate(comb_values):
        b_table[comb[0][comb[1]]] = b_calc[idx]
    
    return pred_output, primary_table, secondary_table, b_table
```
